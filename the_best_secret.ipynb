Вот его текст:

"Как я уже писал ранее, данные, с которыми мы работаем, обезличены, то есть реальные id пользователей заменены случайным числом. Анонимизация пользовательских данных - очень важная тема для Data Science. С одной стороны мы хотим знать о поведении наших пользователей как можно больше, с другой стороны нам важно соблюдать приватность наших пользователей.

Поэтому данные о поведении пользователей анонимизируют не только когда речь идет об открытых данных, но и при работе с данными внутри компании. Не заметили сколько писем вам недавно пришло об изменении политики конфеденциальности, хотите узнать почему? В комментарии приложена ссылка на очень интересный доклад об этой проблеме.

А пока что вот вам хакерская задача, за каким вымышленным id скрывается Анатолий Карпов - автор курса, данные которого мы анализируем?

Введите id Анатолия Карпова, под которым он фигурирует в [данных](https://stepik.org/media/attachments/course/4852/submissions_data_train.zip). "

Посмотрим на данные.

```python
import pandas as pd

submissions_data = pd.read_csv('https://stepik.org/media/attachments/course/4852/submissions_data_train.zip')  # выгружаем данные
submissions_data
```

Можно сделать вывод, что:

step_id - id стэпа

timestamp - время отправки решения в формате unix date

submission_status - статус решения

user_id - анонимизированный id юзера

т.к Анатолий автор курса, то он знает правильные ответы и, следовательно, находиться в числе первых, кто решил все задания (или тех, кто правильно решил больше всех заданий).

Получим этот список.

```python
# группируем по id правильно решивших пользователей, считаем правильные ответы для каждого пользователя и сортируем по убыванию 
submissions_data[submissions_data.submission_status == 'correct'] \
                .groupby('user_id') \
                .agg({'submission_status':'count'}) \
                .sort_values(by=['submission_status'], ascending=False)
```

И первый id в топе - 1046. Вводим его в поле для ответов и получаем вот такое сообщение.
